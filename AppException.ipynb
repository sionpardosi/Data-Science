```python
import pandas as pd

# Membaca dataset Data_Mahasiswa.csv dengan pemisah titik koma
data = pd.read_csv('Data_Mahasiswa.csv', sep=';')

# Menghapus spasi ekstra pada nama kolom agar konsisten
data.columns = data.columns.str.strip()

# Contoh 1: Menangani error pada konversi kolom 'NIM' ke tipe numerik
try:
    # Mengonversi kolom 'NIM' menjadi numerik, jika ada data yang tidak valid, akan menimbulkan ValueError
    data['NIM'] = pd.to_numeric(data['NIM'], errors='raise')
except ValueError as ve:
    print("Terjadi error dalam konversi NIM ke numerik:", ve)
finally:
    print("Proses konversi kolom NIM selesai.")

# Contoh 2: Simulasi operasi pembagian untuk menghitung fitur baru
# Misalnya, kita ingin membuat fitur 'Rata_NIM' dengan membagi nilai NIM dengan 'Angkatan Mahasiswa'
# Perlu diperhatikan bahwa jika 'Angkatan Mahasiswa' bernilai nol, maka akan terjadi ZeroDivisionError
def hitung_rata_nim(df):
    try:
        # Konversi kolom 'Angkatan Mahasiswa' ke numerik
        df['Angkatan Mahasiswa'] = pd.to_numeric(df['Angkatan Mahasiswa'], errors='coerce')
        # Operasi pembagian untuk menghasilkan fitur baru
        df['Rata_NIM'] = df['NIM'] / df['Angkatan Mahasiswa']
    except ZeroDivisionError as ze:
        print("Error: Pembagian dengan nol tidak diperbolehkan:", ze)
    except Exception as e:
        print("Terjadi error lainnya:", e)
    else:
        print("Perhitungan Rata_NIM berhasil tanpa error.")
    finally:
        print("Proses perhitungan Rata_NIM telah selesai.")
    return df

# Penerapan fungsi hitung_rata_nim pada dataset
data = hitung_rata_nim(data)

# Menampilkan 5 baris pertama untuk verifikasi
print("\nData mahasiswa setelah perhitungan Rata_NIM:")
print(data.head())
```
