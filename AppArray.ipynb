```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Membaca dataset "Data_Mahasiswa.csv" dengan pemisah titik koma
data = pd.read_csv('Data_Mahasiswa.csv', sep=';')

# Menghapus spasi berlebih pada nama kolom agar konsisten
data.columns = data.columns.str.strip()

# Misalkan kita ingin menganalisis kolom 'Angkatan Mahasiswa'
# Pastikan kolom 'Angkatan Mahasiswa' dikonversi ke tipe numerik
data['Angkatan Mahasiswa'] = pd.to_numeric(data['Angkatan Mahasiswa'], errors='coerce')

# Mengonversi kolom 'Angkatan Mahasiswa' ke array NumPy
angkatan_array = data['Angkatan Mahasiswa'].to_numpy()

# Contoh operasi aritmatika elemen per elemen: menambahkan 1 ke setiap nilai (misalnya, untuk mengkoreksi data)
angkatan_koreksi = angkatan_array + 1

# Operasi statistik: menghitung rata-rata dan standar deviasi dari angkatan
rata_rata = np.mean(angkatan_array)
std_dev = np.std(angkatan_array)

print("Array Angkatan Mahasiswa:")
print(angkatan_array)
print("\nSetiap elemen setelah ditambah 1:")
print(angkatan_koreksi)
print(f"\nRata-rata Angkatan: {rata_rata}")
print(f"Standard Deviation Angkatan: {std_dev}")

# Visualisasi sederhana: Histogram distribusi nilai angkatan
plt.figure(figsize=(6, 4))
plt.hist(angkatan_array, bins=10, color='skyblue', edgecolor='black')
plt.title('Distribusi Angkatan Mahasiswa')
plt.xlabel('Angkatan')
plt.ylabel('Frekuensi')
plt.show()
```
